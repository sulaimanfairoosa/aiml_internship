```python
def sum_numbers(*args):
        """Sums any number of arguments"""   
        total = 0   
        for num in args:       
             total += num   
             return total 
# Using variable arguments
print(sum_numbers(1, 2, 3))   # 6       
print(sum_numbers(10, 20, 30, 40)) # 100
print(sum_numbers(5))        # 5        
```

```python
def create_person(**kwargs):   
    """Creates a person dictionary with given attributes"""   
    person = {}  
    for key, value in kwargs.items():      
        person[key] = value  
    return person
# Using keyword arguments 
person1 = create_person(name="Alice", age=25, city="New York") 
person2 = create_person(name="Bob", occupation="Engineer")
print(person1)
print(person2)
```

```python
# Example 1: Factorial using recursion 
def factorial(n):   
    if n == 0 or n == 1:   # Base case       
      return 1   
    else:                  # Recursive case        
        return n * factorial(n - 1) 
print(factorial(5))  # 120
```

```python
# Example 2: Fibonacci using recursion 
def fibonacci(n):    
    if n == 0:   # Base case        
        return 0    
    elif n == 1: # Base case        
        return 1    
    else:        # Recursive case        
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(6))  # 8        
```

```python
# Example 3: Recursive sum of a list 
def recursive_sum(lst):    
    if not lst:  # Base case (empty list)        
        return 0    
    else:        # Recursive case        
        return lst[0] + recursive_sum(lst[1:]) 
print(recursive_sum([1, 2, 3, 4, 5]))  # 15
```

```python
counter = 0 
def increment():    
    global counter    
    counter += 1    
    print("Counter inside function:", counter) 
    increment() 
    increment() 
    print("Counter outside function:", counter)
```

```python
# 5. Function to compute factorial of a number
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print(factorial(5))
```

```python
# 9. Function to count vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    return count

print(count_vowels("Python"))
```

```python
# 10. Function that checks if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

```python
# 16. Function that finds the GCD of two numbers
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(12, 18))
```

```python
# 18. Function to count the number of words in a string
def word_count(s):
    words = s.split()
    return len(words)

print(word_count("Python is easy"))
```

```python
# 19. Function that accepts a number and returns the sum of its digits
def digit_sum(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

print(digit_sum(1234))
```

```python
# 20. Function to check if a given year is a leap year
def leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

print(leap_year(2024))
```
